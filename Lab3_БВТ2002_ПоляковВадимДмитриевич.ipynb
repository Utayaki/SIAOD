#### Алгоритм Кнута-Морриса-Пратта

```python
#сложности алгоритмов, когда нужно какой-то из них применять
#Префикс функция строится за P + T
#P - длина образца; Т - длина цепочки
#Проход цикла по строке содержит T итераций
#Итого - O(P+T)

#Лучше использовать тогда когда неудаче предшествует череда совпадений
#Таким образом образ сдвигается на большое кол-во элементов

def func_prefix(s):
    l = len(s)
    P = [0]*l
    i, j = 0, 1
    
    while j < l :
        if ord(s[i]) == ord(s[j]):
            P[j] = i + 1
            i += 1
            j += 1
        elif i:
            i = P[i - 1]
        else:
            P[j] = 0
            j += 1
    return P

def kmp(text, sub):
    sub_len = len(sub)
    text_len = len(text)
    
    print(text_len, sub_len)
    
    if not text_len or sub_len > text_len:
        return []
    
    P = func_prefix(sub)
    
    print(">>>", P)
    
    entries = []
    i = j = 0
    
    while i < text_len and j < sub_len:
        if ord(text[i]) == ord(sub[j]):
            if j == sub_len - 1:
                entries.append(i - sub_len + 1)
                j = 0
            else:
                j += 1
            i += 1
        # text[i] 1= sub[j]
        elif j:     # j > 0
            j = P[j-1]
        else:
            i += 1
    return entries

s = "abcabcabdaeab cabCabdrabcabd tuyeabcabdwrt"
sub = "abcabd"

P = kmp(s, sub)

print(P)
for i in P:
    print(s[i:i + len(sub)])
```

```python
from heapq import heappop, heappush
from math import sqrt
class Board:
    def __init__(self, board, history = None):
        if history is None:
            history = []
        self.board = board
        self.side_len = int(sqrt(len(board)))
        self.history = history
    def get_neighbours(self):
        neighs = [processor(self.board.index(0)) for processor in self._get_neighs_processors()]
        self.history.append(self)
        return [Board(state, self.history) for state in neighs if state]
    def _get_neighs_processors(self):
        return tuple([self._left_neighs, self._right_neighs, self._bot_neighs, self._top_neighs])
    def _left_neighs(self, zero_index):
        if zero_index > 0:
            if self._heuristic(zero_index-1, zero_index) == 1:
                return self._calculate_new_state(zero_index, zero_index-1)
        return []
    def _right_neighs(self, zero_index):
        if zero_index < len(self.board)-1:
            if self._heuristic(zero_index+1, zero_index) == 1:
                return self._calculate_new_state(zero_index+1, zero_index)
        return []
    def _bot_neighs(self, zero_index):
        if zero_index+self.side_len < len(self.board):
            if self._heuristic(zero_index+self.side_len, zero_index) == 1:
                return self._calculate_new_state(zero_index+self.side_len, zero_index)
        return []
    def _top_neighs(self, zero_index):
        if zero_index-self.side_len+1 > 0:
            if self._heuristic(zero_index-self.side_len, zero_index) == 1:
                return self._calculate_new_state(zero_index-self.side_len, zero_index)
        return []
    def _calculate_new_state(self, i, j):
        new_state = self.board.copy()
        new_state[i], new_state[j] = new_state[j], new_state[i]
        return new_state
    def __lt__(self, board):
        return self._board_manhattan() < board._board_manhattan()
    def _board_manhattan(self):
        res = 0
        for current in range(len(self.board)):
            target = (self.board[current]-1) % len(self.board)
            res += self._heuristic(current, target)
        return res
    def _heuristic(self, current, target):
        s = self.side_len
        return abs(target % s - current % s) + abs(target // s - current // s)
```

```python
def solution(board):
    node_hash = {}
    chains_queue = []
    target = tuple([*[i for i in range(1, 16)], 0])
    heappush(chains_queue, board)
    while chains_queue:
        cur_chain = heappop(chains_queue)
        cur_node = tuple(cur_chain.board)
        if cur_node == target:
            return cur_chain
        node_hash[cur_node] = len(cur_chain.history)
        for chain in cur_chain.get_neighbours():
            key = tuple(chain.board)
            if key in node_hash:
                if len(chain.history) >= node_hash[key]:
                    continue
                node_hash[key] = len(chain.history)
            heappush(chains_queue, chain)
```
