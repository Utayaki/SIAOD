# Написать следующие виды сортировок

```python
def selection_sort(matrix): #сортировка выбором
    for row in matrix:
        for i in range(len(row)):
            min_idx = i
            for idx in range(i + 1, len(row)):
                if row[idx] < row[min_idx]:
                    min_idx = idx #запоминаем минимальный элемент
                    
            row[i], row[min_idx] = row[min_idx], row[i] #переносим его уже к отсортированным
            
    return matrix
```

```python
def insertion_sort(matrix): #сортировка вставками
    for row in matrix:
        for i in range(1, len(row)):
            key = row[i]
            j = i-1

            while row[j] > key and j >= 0:
                row[j+1] = row[j]
                j -= 1

            row[j+1] = key
        
    return matrix
```

```python
def bubble_sort(matrix): #пузырьковая сортировка/сортировка обменом
    for row in matrix:
        # swapped остается True, если еще осталось, что сортировать
        # мы определяем его в первый раз для проведения первой итерации
        swapped = True
        
        while swapped:
            swapped = False
            
            for i in range(len(row) - 1):
                if row[i] > row[i + 1]: #если это условие нигде не выполняется swapped остается False и цикл останавливается
                    # Меняем элементы
                    row[i], row[i + 1] = row[i + 1], row[i]
                    # Приравниваем к True для проведения очередной итерации
                    swapped = True
    return matrix
```

```python
def Shell_sort(matrix): #сортировка Шелла (сортировка вставками на спидах)
    for row in matrix:
        row_length = len(row)
        gap = row_length//2
        while gap > 0:            
            for iIndex in range(gap, row_length):
                temp = row[iIndex]
                jIndex = iIndex
                while jIndex >= gap and row[jIndex - gap] > temp:
                    row[jIndex] = row[jIndex - gap]
                    jIndex -= gap
                row[jIndex] = temp
            gap //= 2
    return matrix
```

```python
def pyramid_tournament_sort(matrix):#сортировка турнирная
    class Heap:
        def __init__(self, array):
            self.container = array.copy()
            self.sorted = []

        def sort(self):
            while self.container:
                n = len(self.container)
                self.container[0], self.container[n-1] = self.container[n-1], self.container[0]
                self.sorted.append(self.container[n-1])
                self.container.pop(n-1)

                self.heapify(0)

        def heapify(self, i):
            largest = i

            left = 2*i + 1
            right = 2*i + 2

            if left < len(self.container) and self.container[left] > self.container[largest]:
                largest = left
            if right < len(self.container) and self.container[right] > self.container[largest]:
                largest = right

            if largest != i:
                self.container[i], self.container[largest] = self.container[largest], self.container[i]

                self.heapify(largest)

        def build_heap(self):
            n = len(self.container)
            for i in range(n//2 - 1, -1, -1):
                self.heapify(i)
                
    i = 0
    for row in matrix:
        heap = Heap(row)
        heap.build_heap()
        heap.sort()
        
        matrix[i] = list(reversed(heap.sorted))
        
        i += 1
        
    return matrix
```

```python
def quick_sort(matrix): 
    def row_quick_sort(row): #Быстрая сортировка Хоара
        if len(row) <= 1:
            return row
        else:
            q = random.choice(row)
            L = []
            M = []
            R = []
            for elem in row:
                if elem < q:
                    L.append(elem) 
                elif elem > q: 
                    R.append(elem) 
                else: 
                    M.append(elem)
            return row_quick_sort(L) + M + row_quick_sort(R)
        
    i = 0
    for row in matrix:
        matrix[i] = row_quick_sort(row)
        
        i += 1
        
    return matrix
```
