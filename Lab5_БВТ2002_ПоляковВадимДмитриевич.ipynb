```python
#Снежинка Коха (Три кривых Коха)

import turtle
 
size = 300
depth = 3

turtle.speed(1000)
turtle.hideturtle()

turtle.penup()
turtle.goto(-225, 100) 
turtle.pendown()

def koch_curve(size, depth):
    if depth == 0:
        turtle.forward(size)
    else:
        koch_curve(size / 3, depth - 1)
        turtle.left(60)
        koch_curve(size / 3, depth - 1)
        turtle.right(120)
        koch_curve(size / 3, depth - 1)
        turtle.left(60)
        koch_curve(size / 3, depth - 1)
 
 
def draw_koch_snowflake(size, depth):
    for i in range(3):
        koch_curve(size, depth)
        turtle.right(120)
 
 
draw_koch_snowflake(size, depth)
```

```python
#Кривая Гильберта

import turtle

size = 10
depth = 8

turtle.speed(1000)
turtle.hideturtle()

turtle.penup()
turtle.goto(-225, -225) 
turtle.pendown()

def hilbert_curve(size, parity, depth):

    if depth < 1:
        return

    turtle.left(parity * 90)
    hilbert_curve(size, - parity, depth - 1)
    turtle.forward(size)
    turtle.right(parity * 90)
    hilbert_curve(size, parity, depth - 1)
    turtle.forward(size)
    hilbert_curve(size, parity, depth - 1)
    turtle.right(parity * 90)
    turtle.forward(size)
    hilbert_curve(size, - parity, depth - 1)
    turtle.left(parity * 90)

hilbert_curve(size, 1, depth)
```

```python
#Кривая Серпинского

import turtle

size = 4
depth = 3

turtle.speed(1000)
turtle.hideturtle()

turtle.penup()
turtle.goto(-225, 0) 
turtle.pendown()

def half_sierpinski(size, depth):
    if depth == 0:
        turtle.forward(size)
    else:
        half_sierpinski(size, depth-1)
        turtle.left(45)
        turtle.forward(size*(2)**0.5)
        turtle.left(45)
        half_sierpinski(size, depth-1)
        turtle.right(90)
        turtle.forward(size)
        turtle.right(90)
        half_sierpinski(size, depth-1)
        turtle.left(45)
        turtle.forward(size*(2)**0.5)
        turtle.left(45)
        half_sierpinski(size, depth-1)
        
    return

def sierpinski_curve(size, depth):
    half_sierpinski(size, depth)
    turtle.right(90)
    turtle.forward(size)
    turtle.right(90)
    half_sierpinski(size, depth)
    turtle.right(90)
    turtle.forward(size)
    turtle.right(90)
    
    return

sierpinski_curve(size, depth)
```

```python
#Салфетка Серпинского

import turtle

size = 200
level = 5

turtle.speed(1000)
turtle.hideturtle()

turtle.penup()
turtle.goto(-225, -225) 
turtle.pendown()

def sierpinski_triangle(side, level):
    if level == 1:
        for i in range(3):
            turtle.forward(side)
            turtle.left(120)
    else:
        sierpinski_triangle(side/2, level-1)
        turtle.penup()
        turtle.forward(side/2)
        turtle.pendown()
        sierpinski_triangle(side/2, level-1)
        turtle.penup()
        turtle.back(side/2)
        turtle.left(60)
        turtle.forward(side/2)
        turtle.right(60)
        turtle.pendown()
        sierpinski_triangle(side/2, level-1)
        turtle.penup()
        turtle.left(60)
        turtle.back(side/2)
        turtle.right(60)
        turtle.pendown()
        
sierpinski_triangle(size, level)
```

```python
#Ковер Серпинского

import turtle

size = 400
depth = 3

turtle.speed(1000)
turtle.hideturtle()

turtle.penup()
turtle.goto(-225, -225) 
turtle.pendown()

def sierpinski_carpet(size, depth):
    if depth == 0:
        turtle.color('black')
        turtle.begin_fill()
        for i in range(4):
            turtle.forward(size)
            turtle.left(90)
        turtle.end_fill()

    else:
        for i in range(4):
            sierpinski_carpet(size/3, depth-1)    
            turtle.forward(size/3)

            sierpinski_carpet(size/3, depth-1)    
            turtle.forward(size/3)

            turtle.forward(size/3)
            turtle.left(90)

    return

sierpinski_carpet(size, depth)
```

```python
import turtle  
import math  

turtle.speed(1000)
turtle.hideturtle()

depth = 6

def pythagorean_tree(aturt, depth, maxdepth):  
    if depth > maxdepth:  
        return  
    length = 180*((math.sqrt(2)/2)**depth)  
    anotherturt = aturt.clone()  
    aturt.forward(length)  
    aturt.left(45)  
    pythagorean_tree(aturt, depth+1, maxdepth)  
    anotherturt.right(90)  
    anotherturt.forward(length)  
    anotherturt.left(90)
    anotherturt.forward(length)
    if depth != maxdepth:
        turt3 = anotherturt.clone()  
        turt3.left(45)  
        turt3.forward(180*((math.sqrt(2)/2)**(1+depth)))  
        turt3.right(90)  
        pythagorean_tree(turt3, depth+1, maxdepth)  
    anotherturt.left(90)  
    anotherturt.forward(length)  
    
    
def draw_pythagorean_tree(depth):   
    turt = turtle.Turtle()
    turt.speed(1000)
    #turt.hideturtle()
    turt.penup()  
    turt.goto(-75, -225)  
    turt.pendown()   
    turt.left(90)  
    pythagorean_tree(turt, 1, depth)  


draw_pythagorean_tree(depth)
```

```python
#Таблицы с временем рисования фракталов в зависимости от глубины итерации
import time
import numpy as np
import turtle

turtle.speed(1000)
turtle.hideturtle()

turtle.penup()
turtle.goto(0, 0) 
turtle.pendown()

fractals_time = [[None for i in range(6)] for j in range(6)]
for i in range(6):
    turtle.penup()
    turtle.goto(0, 0) 
    turtle.pendown()
    start_time = time.time()
    draw_koch_snowflake(300, i)
    fractals_time[0][i] = time.time() - start_time

for i in range(6):
    turtle.penup()
    turtle.goto(0, 0) 
    turtle.pendown()
    start_time = time.time()
    hilbert_curve(10, 1, i)
    fractals_time[1][i] = time.time() - start_time
    
for i in range(6):
    turtle.penup()
    turtle.goto(0, 0) 
    turtle.pendown()
    start_time = time.time()
    sierpinski_curve(4, i)
    fractals_time[2][i] = time.time() - start_time

fractals_time[3][0] = 0
for i in range(1, 6):
    turtle.penup()
    turtle.goto(0, 0) 
    turtle.pendown()
    start_time = time.time()
    sierpinski_triangle(200, i)
    fractals_time[3][i] = time.time() - start_time

fractals_time[4][0] = 0
for i in range(1, 4):
    turtle.penup()
    turtle.goto(0, 0) 
    turtle.pendown()
    start_time = time.time()
    sierpinski_carpet(300, i)
    fractals_time[4][i] = time.time() - start_time

fractals_time[5][0] = 0
for i in range(1, 6):
    turtle.penup()
    turtle.goto(0, 0) 
    turtle.pendown()
    start_time = time.time()
    draw_pythagorean_tree(i)
    fractals_time[5][i] = time.time() - start_time
    
    
    
np.save("fractals_time.npy", np.array(fractals_time))
```
