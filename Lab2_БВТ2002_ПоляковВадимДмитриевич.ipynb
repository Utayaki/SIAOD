```python
#Бинарное дерево
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
        
    def bintree_insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.bintree_insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.bintree_insert(data)
        else:
            self.data = data
            
    def bintree_print(self):
        if self.left:
            self.left.bintree_print()
            
        #print(self.data),
        if self.right:
            self.right.bintree_print()
            
    def search(self, value, parent):
        if self is None:
            return -1
        elif value == self.data:
            return [self,parent]
        elif value < self.data:
            return self.left.search(value, self)
        else:
            return self.right.search(value, self)
        
    def bintree_remove(self, data):
        b = self.search(data,None)
        a = b[0]
        
        if a is not None:
            if a.left is not None and a.right is None:
                if b[1].left == a:
                    b[1].left = a.left
                if b[1].right == a:
                    b[1].right = a.left
                a.left = a.left.left
                a.right = a.left.right
                a.data = a.left.data
                a = a.left
                
            elif a.right is not None and a.left is None:
                if b[1].left == a:
                    b[1].left = a.right
                if b[1].right == a:
                    b[1].right = a.right
                a.left = a.right.left
                a.right = a.right.right
                a.data = a.right.data
                a = a.right
                
            elif a.left is not None and a.right is not None:
                y = a
                while y.left is not None:
                    x = y
                    y = y.left
                if x != a and y.right is not None:
                    y.left = y.right.left
                    y.right = y.right.right
                    y.data = y.right.data
                    y = y.right
                    y.right = a.right
                a.left = y.left
                a.right = y.right
                a.data = y.data
                a = y
                y.left = a.left
            else:
                if b[1].left == a:
                    b[1].left = None
                if b[1].right == a:
                    b[1].right = None
                a = None
                
    def bintree_search(self, value):
        if self is None:
            return -1
        elif value == self.data:
            return self.data
        elif value < self.data:
            return self.left.bintree_search(value)
        else:
            return self.right.bintree_search(value)
        
root = Node(10)

for i in range(10):
    root.bintree_insert(i)
    
root.bintree_print()
root.bintree_remove(6)
print("|")
root.bintree_print()

root.bintree_search(2)
```

```python
#Интерполяционный
#В среднем O(log(log(n))) (значения распределены равномерно)
def InterpolationSearch(lys, val):
    low = 0
    high = (len(lys) - 1)
    
    while low <= high and val >= lys[low] and val <= lys[high]:
        
        index = low + int(((float(high - low)/(lys[high] - lys[low])) * (val - lys[low])))
        
        if lys[index] == val:
            return index
        
        if lys[index] < val:
            low = index + 1;
            
        else:
            high = index - 1;
            
    return -1
```

```python
#Простое рехеширование
class HashTable:
    def __init__(self, n = 5):
        self.size = 0
        self.slots = [None] * n
        self.data = [None] * n
        self.q = 0

    def __setitem__(self, key, value):
        hashvalue = self._hash(key)
        if self.slots[hashvalue] is None:
            self.slots[hashvalue] = key
            self.data[hashvalue] = value
        else:
            if self.slots[hashvalue] == key:
                self.data[hashvalue] = value
            else:
                nextslot = self._rehash(hashvalue)
                self.q += 1
                while self.slots[nextslot] is not None and self.slots[nextslot] != key:
                    nextslot = self._rehash(nextslot)
                    self.q += 1
                if self.slots[nextslot] is None:
                    self.slots[nextslot] = key
                    self.data[nextslot] = value
                else:
                    self.data[nextslot] = value
        self.q = 0
        
    def __getitem__(self, key):
        startslot = self._hash(key)
        position = startslot
        while self.slots[position] is not None:
            if self.slots[position] == key:
                self.q = 0
                return self.data[position]
            else:
                position = self._rehash(position)
                self.q += 1
                if position == startslot:
                    continue
        self.q = 0
        
        return -1
    def _hash(self, key):
        return key % len(self.slots)
    
    def _rehash(self, oldhash):
        return (oldhash + 1) % len(self.slots)
```

```python
#Рехэширование с помощью псевдослучайных чисел
import random
random.seed(42)


class RandomHashTable(HashTable):
    def __init__(self, n = 5):
        super().__init__(n)
        
        self.randints = [random.randint(0, 1000) for i in range(n)]
        
    def _rehash(self, oldhash):
        if self.q == len(self.randints):
            self.randints.extend([random.randint(0, 100) for i in range(self.q)])
            
        newhash = (oldhash + self.randints[self.q]) % len(self.slots)
        
        return newhash
```

```python
#Метод цепочек
class ListNode:
    def __init__(self, key, val):
        self.pair = (key, val)
        self.next = None
        
class ChainedHashTable:
    def __init__(self):
        self.m = 1000;
        self.h = [None] * self.m
        
    def __setitem__(self, key, value):
        index = self._hash(key)
        cur = self.h[index]
        
        if cur is None:
            self.h[index] = ListNode(key, value) 
        else:
            while True:
                if cur.pair[0] == key:
                    cur.pair = (key, value)
                    return
                if cur.next == None: break
                cur = cur.next
                
            cur.next = ListNode(key, value)
            
    def __getitem__(self, key):
        index = self._hash(key)
        cur = self.h[index]
        
        while cur:
            if cur.pair[0] == key:
                return cur.pair[1]
            else:
                cur = cur.next
        return -1
    
    def _hash(self, key):
        return key % self.m
```
